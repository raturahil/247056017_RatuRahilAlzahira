```python
import numpy as np
```

**Example 1**

```python
# Inisialisasi parameter
weights = np.array([0.0, 0.0, 0.0])  # [w1, w2, b]
learning_rate = 1
theta = 0
epochs = 10

# Data input
X = np.array([[1, 1],
              [1, 0],
              [0, 1],
              [0, 0]])

# Tambah kolom bias dengan nilai 1
X = np.c_[X, np.ones(X.shape[0])]

# Data target output
T = np.array([1, -1, -1, -1])

# Fungsi aktivasi
def activation(y_in):
    if y_in > theta:
        return 1
    elif y_in < -theta:
        return -1
    else:
        return 0

# Algoritma Perceptron
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    weights_changed = False  # Untuk memeriksa apakah ada perubahan bobot
    
    for i in range(len(X)):
        # Hitung output perceptron
        y_in = np.dot(X[i], weights)
        y = activation(y_in)
        
        # Cek apakah output sama dengan target
        if y != T[i]:
            # Update bobot jika terjadi error
            weights += learning_rate * T[i] * X[i]
            weights_changed = True
            print(f"  Update weights to: {weights}")
    
    # Kondisi berhenti jika tidak ada perubahan bobot
    if not weights_changed:
        print("Training converged.")
        break
    else:
        print(f"  Weights after epoch {epoch+1}: {weights}")

print("Final weights:", weights)
```

**Example 2**

```python
# Inisialisasi parameter
weights = np.array([0.5, 0.5, 0.5])
learning_rate = 1
theta = 0.8
epochs = 10

# Data input
X = np.array([[1, 0, 0],
              [0, 0, 1],
              [1, 0, 1],
              [1, 1, 1]])

# Data target output
T = np.array([0, 0, 0, 1])

# Fungsi aktivasi
def activation(y_in):
    return 1 if y_in >= theta else 0

# Algoritma Perceptron
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    weights_changed = False  # Untuk memeriksa apakah ada perubahan bobot

    for i in range(len(X)):
        # Hitung output perceptron
        y_in = np.dot(X[i], weights)
        y = activation(y_in)
        
        # Cek apakah output sama dengan target
        if y != T[i]:
            # Update bobot jika terjadi error
            weights += learning_rate * (T[i] - y) * X[i]
            weights_changed = True
            print(f"  Update weights to: {weights}")
    
    # Kondisi berhenti jika tidak ada perubahan bobot
    if not weights_changed:
        print("Training converged.")
        break
    else:
        print(f"  Weights after epoch {epoch+1}: {weights}")

print("Final weights:", weights)
```
